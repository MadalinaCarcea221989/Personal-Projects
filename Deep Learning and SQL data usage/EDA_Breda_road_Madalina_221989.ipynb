### Starting the EDA , data cleaning

```python
class DataProcessor:
    def __init__(self, df, selected_columns):
        """
        Initializes the DataProcessor with a DataFrame and selected columns.

        Args:
            df (pd.DataFrame): The DataFrame containing the data.
            selected_columns (list): List of columns to be processed.
        """
        self.df = df
        self.selected_columns = selected_columns
        self.eda_df = None
        self.conn = None
        self.cursor = None
        self.db_params = {}
        self.query = ""
        
        # Configure logging
        logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
        self.logger = logging.getLogger(__name__)

    def connect_db(self):
        """
        Establishes a connection to the database using provided parameters.
        """
        try:
            self.conn = psycopg2.connect(**self.db_params)
            self.cursor = self.conn.cursor()
            self.logger.info("Database connection successful.")
        except (Exception, psycopg2.DatabaseError) as error:
            self.logger.error(f"Error: {error}")
            if self.conn:
                self.conn.rollback()

    def fetch_data(self):
        """
        Fetches data from the database and stores it in the DataFrame.
        """
        try:
            self.cursor.execute(self.query)
            rows = self.cursor.fetchall()
            columns = [desc[0] for desc in self.cursor.description]
            self.df = pd.DataFrame(rows, columns=columns)
            self.logger.info("Data fetched successfully.")
        except (Exception, psycopg2.DatabaseError) as error:
            self.logger.error(f"Error: {error}")
            if self.conn:
                self.conn.rollback()

    def preprocess_data(self):
        """
        Preprocesses the data by handling missing values, removing unwanted columns, and normalizing numeric columns.

        Steps involved:
        1. Replace missing values (including empty strings) with NaN.
        2. Provide initial data overview and missing values summary.
        3. Drop columns that are completely missing or not useful for the project.
        4. Handle partial missing values in numeric columns by filling with median.
        5. Handle outliers and normalize numeric columns.

        Updates `self.df` with the preprocessed data.

        Raises:
            ValueError: If `self.df` is empty after preprocessing.

        Example:
            >>> processor = DataProcessor(df, selected_columns)
            >>> processor.preprocess_data()
        """
        # Replacing missing values (including empty strings) with NaN if necessary
        self.df = self.df.apply(lambda x: x.replace('', np.nan))
        
        # Initial data overview
        self.logger.info("Initial Data Overview:")
        self.logger.info(self.df.describe())
        self.logger.info("Initial Missing Values:")
        self.logger.info(self.df.isnull().sum())
        
        # Drop columns that are completely missing after checking in the previous step
        self.selected_columns = [col for col in self.selected_columns if self.df[col].notna().any()]
        
        # Drop '_id', '_key', and 'bronhouder' columns as they are not useful for the project
        self.selected_columns = [col for col in self.selected_columns if col not in ['_id', '_key', 'bronhouder','inonderzoek','relatievehoogteligging']]
        
        # Preprocess only the selected columns
        self.eda_df = self.df[self.selected_columns].copy()
        
        # Handle partial missing values (filled with median) only for numeric columns
        numeric_cols = self.eda_df.select_dtypes(include=[np.number]).columns
        self.eda_df[numeric_cols] = self.eda_df[numeric_cols].fillna(self.eda_df[numeric_cols].median())
        
        # Normalize only the numeric columns
        if not self.eda_df.empty:
            
            # Handle outliers
            for column in numeric_cols:
                self.handle_outliers(column)
            
            # Normalization
            scaler = MinMaxScaler()
            self.eda_df[numeric_cols] = scaler.fit_transform(self.eda_df[numeric_cols])
        
            self.logger.info("Data preprocessing completed.")
        
            # Merging the preprocessed columns back into the original DataFrame
            self.df.update(self.eda_df)
        else:
            self.logger.error("The DataFrame is empty after handling missing values and outliers.")
            raise ValueError("The DataFrame is empty after handling missing values and outliers.")
    
    def handle_outliers(self, column):
        """
        Handles outliers for the specified column by capping them at the 1st and 99th percentiles.

        Args:
            column (str): The name of the column to handle outliers for.
        """
        lower_bound = self.eda_df[column].quantile(0.01)
        upper_bound = self.eda_df[column].quantile(0.99)
        self.eda_df[column] = np.clip(self.eda_df[column], lower_bound, upper_bound)
        self.logger.info(f"Outliers handled for column: {column}")
    
    def split_data(self):
        """
        Splits the data into training, validation, and test sets.

        Returns:
            tuple: DataFrames for training, validation, and test sets.
        """
        if not self.eda_df.empty:
            train_df = self.eda_df.sample(frac=0.7, random_state=42)
            temp_df = self.eda_df.drop(train_df.index)
            validation_df = temp_df.sample(frac=0.5, random_state=42)
            test_df = temp_df.drop(validation_df.index)
            return train_df, validation_df, test_df
        else:
            return None, None, None
    
    def perform_eda(self):
        """
        Performs Exploratory Data Analysis (EDA) on the processed data.
        """
        if not self.eda_df.empty:
            self.logger.info("\nTable:")
            display(self.eda_df)
            
            self.logger.info("\nDescriptive Statistics:")
            self.logger.info(self.eda_df.describe())
            
            self.logger.info("\nMissing Values:")
            self.logger.info(self.eda_df.isnull().sum())
            
        else:
            self.logger.error("EDA cannot be performed on the empty DataFrame.")
    
    def save_to_csv(self, filename):
        """
        Saves the processed DataFrame to a CSV file.

        Args:
            filename (str): The path to the file where the DataFrame will be saved.
        """
        if not self.eda_df.empty:
            self.eda_df.to_csv(filename, index=False)
            self.logger.info(f"\nDataFrame saved to '{filename}'")
        else:
            self.logger.error("No data available to save to CSV.")
    
    def document_preprocessing(self, filename):
        """
        Documents the preprocessing steps in a file.

        Args:
            filename (str): The path to the file where the documentation will be saved.
        """
        with open(filename, 'w') as f:
            f.write("Documenting the preprocessing steps in here...")
        self.logger.info(f"\nDocumentation saved to '{filename}'")
    
    def close_connection(self):
        """
        Closes the database connection.
        """
        if self.cursor:
            self.cursor.close()
        if self.conn:
            self.conn.close()
        self.logger.info("Database connection closed.")

# Database connection parameters and query
db_params = {
    'host': '194.171.191.226',
    'port': '6379',
    'database': 'postgres',
    'user': 'group23',
    'password': 'blockd_2024group23_14'
}
query = "SELECT * FROM data_lake.breda_road"
selected_columns = ['objecteindtijd', 'eindregistratie', 'wegdeeloptalud', '_id', '_key', 'bronhouder','inonderzoek','relatievehoogteligging']

# Initialize the DataFrame (if already available)
df = pd.DataFrame()  # Placeholder, replace with actual DataFrame if available

# Create processor object and process data
processor = DataProcessor(df, selected_columns)
processor.db_params = db_params
processor.query = query
processor.connect_db()
processor.fetch_data()

# Update selected columns to include all columns except those completely missing (from the selected_columns)
processor.selected_columns = [col for col in processor.df.columns if processor.df[col].notna().any()]
processor.preprocess_data()

# Splitting the dataset
train_df, validation_df, test_df = processor.split_data()

# Processing functions being used
processor.perform_eda()
processor.save_to_csv('./columns_second.csv')
processor.close_connection()
```
